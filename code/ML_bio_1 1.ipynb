```python
# Calculate the correlation matrix
corr_matrix = df1.corr()

# Set the threshold
threshold = 0.6

# Find pairs of variables with correlation above the threshold
high_corr_pairs = []

# Loop through the correlation matrix
for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if abs(corr_matrix.iloc[i, j]) > threshold:
            pair = (corr_matrix.columns[i], corr_matrix.columns[j], corr_matrix.iloc[i, j])
            high_corr_pairs.append(pair)

# Create a DataFrame from the list of high correlation pairs
high_corr_df = pd.DataFrame(high_corr_pairs, columns=['Variable 1', 'Variable 2', 'Correlation'])
```

# Handling Outliers

```python
for col in df1.columns:
    print(col)
    sns.boxplot(data=df1, x=df1[col])
    plt.show()
```

Series:- A pandas Series is a one-dimensional array-like object that can hold various types of data, such as integers, floats, strings, etc., and it is essentially a column in a pandas DataFrame.
series: The pandas Series (a column from the DataFrame) to be Winsorized.
series.quantile(low): Computes the value at the low percentile (e.g., 5th percentile).
series.quantile(high): Computes the value at the high percentile (e.g., 95th 

```python
# Outlier Treatment - Winsorization
def winsorize_series(series, low=0.05, high=0.95):
    return np.clip(series, series.quantile(low), series.quantile(high))

for col in df1.columns:
    df1[col] = winsorize_series(df1[col])
```

# Handling Skeweness

```python
for col in df1.columns:
    skewness = df1[col].skew()
    sns.displot(df1[col].skew())
    print(f" {col} :  {skewness}")
```

```python
d={}
for col in df1.columns:
    skewness = df1[col].skew()
    d[col] = skewness
    print(f" {col} :  {skewness}")
```

```python
df_skew = pd.DataFrame(list(d.items()), columns=['Column', 'Skewness'])
print(df_skew)
```

```python
sk=df_skew[(df_skew['Skewness']>1)|(df_skew['Skewness']<-1)].Column.tolist()
sk
```

```python
import numpy as np

# Assume 'sk' is a list of columns to transform
for col in sk:
    if col in df1.columns:
        if df1[col].skew() > 1:
            df1[col] = np.log(df1[col])
        elif df1[col].skew() < -1:
            df1[col] = np.square(df1[col])
```

```python
# Check Skewness again
skewness_after = df1.skew()
print("Skewness after transformation:\n", skewness_after)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
xtrain = sc.fit_transform(xtrain)
xtrain
```

```python
xtest = sc.transform(xtest)
xtest
```

```python
def mymodel(model):
    model = model.fit(xtrain,ytrain)
    ypred = model.predict(xtest)
    print(ypred)
    xpred = model.predict(xtrain)
    print(xpred)
    fig = plt.figure(figsize=(20,10))
    ax1 = fig.add_subplot(2,2,1)
    ax1.scatter(ytest,ypred,color="blue")
    ax1.grid()
    ax1.set_title("Testing prediction plot")

    ax2 = fig.add_subplot(2,2,2)
    ax2.scatter(ytrain,xpred,color="red")
    ax2.grid()
    ax2.set_title("Training prediction plot")
    plt.show()

    train_score = model.score(xtrain,ytrain)
    print(f"train_score: {train_score}")
    test_score = model.score(xtest,ytest)
    print(f"test_score: {test_score}")

    mse = mean_squared_error(ytest,ypred)
    print(f"mse : {mse}")
    return model
```

```python
linreg = LinearRegression()
```

```python
res = mymodel(linreg)
```
